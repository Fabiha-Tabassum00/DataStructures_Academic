```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

```python
def landscape(m):
  mat = np.zeros((m.shape[1],m.shape[0]), dtype=int)
  r = len(m[0])
  c = len(m)
  for row in range(r):
    for col in range(c):
      mat[row][col] = m[col][row]
  return mat


m = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(landscape(m))
```

```python
def show_knight_move(knight):
  b = np.zeros((8, 8), dtype=int)
  x1,x2 = knight
  moves = [
        (-2, -1), (-2, 1), (2, -1), (2, 1),
        (-1, -2), (-1, 2), (1, -2), (1, 2)
  ]
  for i in moves:
    x3, x4 = i
    new_x, new_y = x1 + x3, x2 + x4
    if 0<=new_x< 8 and 0<=new_y< 8:
      b[new_x, new_y] = 3
  b[x1, x2] = 33
  return b


knight = (3,4)
chess_board = show_knight_move(knight)
print_matrix(chess_board)
```
